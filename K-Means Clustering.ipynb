```python
df.head()   # We have categorical features and we cannot use them for clustering
```

```python
X = pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

```python
from sklearn.cluster import KMeans
```

```python
help(KMeans)
```

```python
# For our model, there could be clusters - people who can be contacted for further campaigns and those who cannot
model = KMeans(n_clusters=2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels    # All rows are grouped into 2 labels - 0 or 1
# 0 and 1 does not mean anything - just 2 groups.
# It may switch 0s and 1s if we run it again
```

```python
X['Cluster'] = cluster_labels
```

```python
X
```

```python
X.corr()['Cluster']    
# Notice that we didn't make drop_first = True 
# So we can see mirror images - subscribed_no and subscribed_yes are the same with different signs
```

```python
X.corr()['Cluster'][:-1].sort_values()  # We have dropped Cluster corr with itself
```

```python
plt.figure(figsize=(12,6), dpi=100)
X.corr()['Cluster'][:-1].sort_values().plot(kind='bar')
plt.xticks(rotation=90);
```

```python
# Checking how good our K value is
# Consider all your clusters - Measure sum of distances from points to cluster centers
# We can also consider sum of squared distances (ssd)
# As the number of clusters increases, the distance decreases
# If number of points is equal to the number of clusters, ssd will be zero
# So we're not looking for minimum distance
# We have to go for a K value where rate of reduction in SSD begine to decline
# This means adding an extra cluster is not obtaining enough clarity of cluster separation to justify increasing K
```

```python
# Elbow method
# After a point decrease in SSD begins to flatten out
```

```python
ssd = []

for n in range(2,10):
    model = KMeans(n_clusters=n)
    model.fit(scaled_X)
    
    ssd.append(model.inertia_)   # SSD - Point to cluster center
```

```python
ssd
```

```python
plt.plot(range(2,10), ssd, 'o--')
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()
# Gives difference between current row and previous row
# Least transition between index 5 to 6 --> K value 6 to 7
```
